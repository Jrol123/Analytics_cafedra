## Рабочая версия

```python
disk_url = "https://disk.yandex.ru/d/1SQxAcrlg3Kz3g"
mod1_path = "/Модуль%201/Проверка%20заданий%201%20модуль.xlsx"
mod2_path = "/Модуль%202/Проверка%20заданий%202%20модуль.xlsx"
```

```python
def get_url(disk_path: str, file_path: str):
    return rq.get(
        f"https://cloud-api.yandex.net/v1/disk/public/resources/download?public_key={disk_path}&path={file_path}"
    ).json()["href"]
```

```python
mod1_dUrl = get_url(disk_url, mod1_path)
mod1_dUrl
```

```python
mod2_dUrl = get_url(disk_url, mod2_path)
mod2_dUrl
```

## Загрузка

```python
def get_xlsx(name_file, url):
    file_Path = f'data/{name_file}.xlsx'
    file_xlx = rq.get(url)
    with open(file_Path, 'wb') as file:
        file.write(file_xlx.content)
```

```python
get_xlsx("mod1_marks", mod1_dUrl)
get_xlsx("mod2_marks", mod2_dUrl)
```

# Обработка

```python
mod1 = pd.read_excel("data/mod1_marks.xlsx").fillna(0)
mod2 = pd.read_excel("data/mod2_marks.xlsx", sheet_name="Machine Learning").fillna(0)
```

```python
def clean_csv(xlx, attest_col_name):
    clear_xlx = xlx.copy()
    clear_xlx[attest_col_name].replace(to_replace="Незачёт", value=False, inplace=True)
    clear_xlx[attest_col_name].replace(to_replace="Зачёт", value=True, inplace=True)
    clear_xlx[attest_col_name] = clear_xlx[attest_col_name].astype(bool)
    clear_xlx.set_index("ФИО", inplace=True)
    return clear_xlx
```

```python
mod1 = clean_csv(mod1, "Аттестация по модулю")
mod2 = clean_csv(mod2, "Аттестация")
```

```python
mod1.to_csv("data/mod1_marks.csv")
mod2.to_csv("data/mod2_marks.csv")
```
