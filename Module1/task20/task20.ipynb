```python
from datetime import datetime

df['car_age'] = datetime.now().year - df['year']
df['odometer_per_price'] = df['odometer'] / df['price']
df['odometer_kilometer'] = df['odometer'] / 1.609344
df['odometer_kilometer_per_price'] = df['odometer_km']
df['description_word_count'] = df['description'].apply(lambda x: len(str(x).split()))

df
```
